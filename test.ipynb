Necessary imports to run the models:

```python
from transformer_blocks.gpt2 import GPT2 
from transformer_blocks.gpt2_gqa import GQAGPT2
from transformer_blocks.gpt2_moe import MoEGPT2
from gpt_Pretraining.text_generation import Text_Generation
import torch
```

Download and run the text generation:

```python
#Download Stories-SLM 2 model:
model = GQAGPT2.from_pretrained("NamrataThakur/Small_Language_Model_GQA_48M_Pretrained")

model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Model loaded and ready!")

#---------------------------- Checking the generation to make everything is okay ---------------------------

generation = Text_Generation(model=model, device=device, tokenizer_model='gpt2', 
                                          arch_type='GQA') #arch_type = "GQA" for 48M Stories-SLM 2 model,  
                                                          # arch_type = "original" for 53M Stories-SLM model  

print('----------------------- EXAMPLE 1 --------------------------------')
start_context = "Bob and Billy went to the "
print('PROMPT : ', start_context)
response = generation.text_generation(input_text=start_context, max_new_tokens = 160, temp = 0.2, top_k=3, kv_cache=True)
response
```
