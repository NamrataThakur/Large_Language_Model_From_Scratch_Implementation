```python
with open('the-bpe-text.txt', "r", encoding = 'utf-8') as file:
    raw_text = file.read()
    
#Getting the raw bytes as tokens from the text:
byte_tokens = raw_text.encode('utf-8')
print('Raw bytes: ', byte_tokens)

# Converting the tokens to int:
int_tokens = list(map(int, byte_tokens))
print('Raw bytes converted to integer tokens: \n', int_tokens)
print('Token range: {0} - {1}'.format(min(int_tokens), max(int_tokens)))
print('Total number of tokens: ', len(int_tokens))
print('Printing one character from the byte token: ', chr(65))
```

```python
# Step 1: Find the pair of byte (int,int) that are frequently present and store the count of occurences:
def get_most_common(l_tokens,reversed=False):
    frequency = {}
    for pair in zip(l_tokens, l_tokens[1:]):
        frequency[pair] = frequency.get(pair,0) + 1

    # Step 2: Sort the dictionary in the descending/ascending order of the value:
    sorted_frequency_dict = {key : value for key, value in sorted(frequency.items(), 
                                                              key= lambda frequency : frequency[1], 
                                                              reverse= reversed )
                                                              }
    return sorted_frequency_dict

sorted_frequency_dict = get_most_common(int_tokens,True)
print(sorted_frequency_dict)
```

```python
# Step 3: Create a function that will merge the most frequently occuring pair with a new token:
def tokens_merge(token_list, pair, new_token):
    merged_tokens = []
    i = 0
    #for i in range(len(token_list)):
    while i < len(token_list):
        #print('1st i',i)
        if i < len(token_list) -1 and token_list[i] == pair[0] and token_list[i+1] == pair[1]:
            merged_tokens.append(new_token)
            i = i + 2
            #print('2nd i',i)
        else:
            #print(token_list[i])
            merged_tokens.append(token_list[i])
            i = i + 1
    return merged_tokens

print(len([5, 6, 6, 7, 9, 1]))
print(tokens_merge([5, 6, 6, 7, 9, 1], (6, 7), 99))
print(len(tokens_merge([5, 6, 6, 7, 9, 1], (6, 7), 99)))
```

```python
# Step 4: Create the finalized dictionary containing merging pairs and replacing token: 
def merged_replaced_dict(token_list,inital_new_token, merge_times):
    merges = dict()
    merged_tokens = token_list
    sorted_merged_frequency = get_most_common(merged_tokens,True)
    #print(sorted_merged_frequency)
    new_token = inital_new_token
    for i in range(merge_times):
        merged_tokens = tokens_merge(merged_tokens,list(sorted_merged_frequency.keys())[0],new_token )
        merges[list(sorted_merged_frequency.keys())[0]] = new_token
        print(f'Merging {list(sorted_merged_frequency.keys())[0]} with {new_token}. ')
        sorted_merged_frequency = get_most_common(merged_tokens,True)
        #print(sorted_merged_frequency)
        #print(new_token)
        new_token = new_token + 1

    return merges, merged_tokens

inital_new_token = 256 
merge_times = 20
merges, final_tokens = merged_replaced_dict(int_tokens,inital_new_token, merge_times)
print(merges)
print('Length of initial text : ', len(int_tokens))
print('Lenghth of final text: ', len(final_tokens))
print(f"Compression ratio: {len(int_tokens) / len(final_tokens):.2f}X")
```

```python
# Step 5: Create the finalized vocabulary with additional tokens:
def final_vocabulary(merged_dict, vocab_size):
    vocab = dict()
    vocab = {token_id : bytes([token_id]) for token_id in range(vocab_size)}
    for k,v in merged_dict.items():
        vocab[v] = vocab[k[0]] + vocab[k[1]]
    return vocab

global vocab 
vocab = final_vocabulary(merges, inital_new_token)
print('Final Vocabulary: ', vocab)
print('Final Vocab Size : ', len(vocab))
```

```python
# Step 6: Decoding: Given a list of token_id (byte tokens) return the string:
def decode(token_list):
    
    output_tok = b''.join(vocab[tok] for tok in token_list)
    output_string = output_tok.decode('utf-8', errors="replace")
    return output_string

output = decode(final_tokens)
print(output)
```

```python
# Step 7: Encoding: Given the raw text return the merged byte token list:
def encode(text_input):
    tokens = list(map(int, text_input.encode('utf-8')))
    #print(tokens)
    while len(tokens) >= 2:
        sorted_frequency_dict = get_most_common(tokens, False)
        #print(sorted_frequency_dict)
        pair = min(sorted_frequency_dict, key = lambda x: merges.get(x, float("inf")))
        #print(pair)
        if pair not in merges:
            #print("Pair cannot be merged!")
            break
        else:
            new_token = merges[pair]
            tokens = tokens_merge(tokens,pair,new_token)
    return tokens
```
